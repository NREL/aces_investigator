```python
import fiona
import pandas
import pywtk.site_lookup as site_lookup
import pywtk.wtk_api
from shapely import wkb
from shapely.geometry import Point
from shapely.geometry import shape, MultiPolygon
import matplotlib.pyplot as plt
```

```python
# lats and longs. These lats and longs are located in ReEDS regions, from which 
# we will select wind toolkit sites. We associate IDs with the lats and longs.
windGenIDs = [317,309,303,122]
windGenLongsLats = [[-116.3243473,32.81886407],[ -118.0435499,35.21754253]
                    ,[-117.0558295,35.37843341],[-118.127342,34.73575793]]
windGenCapacity = [796.3,148.3,844.1,711.4]
```

```python
# Put the lats, longs, and IDs into a data frame
siteInfoDf = pandas.DataFrame(windGenCapacity,index=windGenIDs,columns=["capacity"])
siteInfoDf["rtsLong"] = [windGenLongsLats[i][0] for i in range(len(windGenIDs))]
siteInfoDf["rtsLat"] = [windGenLongsLats[i][1] for i in range(len(windGenIDs))]
siteInfoDf.index.name = "Plexos Node ID"

# Convert lat longs to hex-encoded wkbs. Do this by putting them in wkt format then encoding
siteInfoDf["the_geom"] = [Point(siteInfoDf["rtsLong"].values[i],siteInfoDf["rtsLat"].values[i]).wkb.encode("hex")
                         for i in range(len(windGenIDs))]
```

```python
# Assigning ReEDSregions to node locations
#pca = fiona.open('../../wind2plexos/ReEDS Shape Files/PCA_Wind_Regions.shp', 'r')
pca = fiona.open('data/ReEDS Shape Files/PCA_Wind_Regions.shp', 'r')

#for entry in pca:
#    # ['geometry', 'type', 'id', 'properties']
#    geom = shape(entry['geometry'])
#    ba = str(entry['properties']['pca'])
#    demreg2 = str(int(entry['properties']['demreg2']))
#    siteInfoDf.loc[lambda df: df['the_geom'].apply(wkb.loads, hex=True).apply(geom.contains), 'region'] = demreg2
```

```python
# Write these site data to csv file
siteInfoDf.to_csv("data/siteInfo.csv")
```

```python
#  Given regions and capacities, select wtk sites from a list sorted by capacity
sites2MakeCapacity = []

# Import three tier sites metadata and sort by annualized capacity
#sites = pandas.read_csv( '../../wind2plexos/three_tier_site_metadata_regions.csv', index_col=0, dtype=object)
sites = pandas.read_csv( 'data/three_tier_site_metadata_regions.csv', index_col=0, dtype=object)
sites['ann_capacity'] = sites['capacity'].astype(float) * sites['capacity_factor'].astype(float)
sites = sites.sort_values('ann_capacity', ascending=False)


for i in range(len(siteInfoDf)):
    region = siteInfoDf['region'].values[i]
    capacity = siteInfoDf['capacity'].values[i]
    
    # gather sites in region
    regionSites = sites.loc[sites.region == region]
    
    sum_cap = 0
    site_count = 0
    
    # select sites 
    for index, row in regionSites.iterrows():
        site_count += 1
        #sum_cap += row['ann_capacity'] #I need different data to do this
        sum_cap += float(row['capacity'])
        if sum_cap >= capacity:
            break
            
    sites2MakeCapacity.append(regionSites[:site_count])
```

```python
# Check how many sites we have per region
map(len,sites2MakeCapacity)
```

```python
# export dataframes for each windfarm
for i in range(len(windGenIDs)):
    sites2MakeCapacity[i].to_csv("data/node_"+str(siteInfoDf.index.values[i])+"_WTK_sites.csv")
```

```python
# try plotting a region

# code snippet from example_plot.ipynb
region_map = {}
for entry in pca:
    # ['geometry', 'type', 'id', 'properties']
    region_map[str(entry['properties']['OBJECTID'])] = shape(entry['geometry'])
mp = MultiPolygon( region_map.values() )
```

```python
# For a region, plot the sites we selected
#Note: difference between "region" and "sites.region" objectid75 corresponds to demreg2 (I need to fix this later)

region = '78'
area = region_map[region]

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
ax.set_aspect(1)
(x, y) = area.exterior.xy
ax.plot(x, y, color='#222222', linewidth=1, solid_capstyle='round', zorder=1)

# Plot wind sites with color based on annualized capacity
r_sites = sites2MakeCapacity[2]
print("%s sites in region %s"%(len(r_sites), region))
cap_cmap = plt.get_cmap('brg')
sax = ax.scatter(r_sites['lng'].astype(float), r_sites['lat'].astype(float), marker='o', 
                 c=r_sites['ann_capacity'].astype(float), s=100, edgecolors='none', cmap=cap_cmap, vmin=0, vmax=10)
cbar = fig.colorbar(sax, )
cbar.set_label('Annualized Capacity (MW)', rotation=90)

# add a point for the plexos node
ax.scatter(siteInfoDf['rtsLong'].iloc[2].astype(float), siteInfoDf['rtsLat'].iloc[2].astype(float), marker='*',
                s=200, c='Black')
plt.show()
```
